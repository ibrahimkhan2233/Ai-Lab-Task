```python
import numpy as np
import matplotlib.pyplot as plt
arr1 = np.arange(10)
print("1D Array from 0 to 9:", arr1)
arr2 = np.arange(1, 10).reshape(3, 3)
print("2D Array (3x3) with values from 1 to 9:\n", arr2)
arr3 = np.random.randint(1, 101, size=(2, 3, 4))
print("3D Array (2x3x4) with random integers from 1 to 100:\n", arr3)
```

```python
second_row = arr2[1]
print("Second row of arr2:", second_row)
element_3rd_row_2nd_col = arr2[2, 1]
print("Element at 3rd row and 2nd column of arr2:", element_3rd_row_2nd_col)
first_layer = arr3[0]
print("Elements at the 0th layer of arr3:\n", first_layer)
```

```python
reshaped_arr1 = arr1.reshape(2, 5)
print("Reshaped 1D array into 2x5 2D array:\n", reshaped_arr1)
```

```python
arr1_plus_10 = arr1 + 10
print("Array arr1 after adding 10 to each element:", arr1_plus_10)
arr2_times_2 = arr2 * 2
print("Array arr2 after multiplying each element by 2:\n", arr2_times_2)
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
plt.plot(x, y)
plt.title("Basic Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```

```python
x_scatter = [5, 7, 8, 7, 2, 17, 2, 9, 4, 11]
y_scatter = [99, 86, 87, 88, 100, 86, 103, 87, 94, 78]

plt.scatter(x_scatter, y_scatter, color='red', s=100)
plt.title("Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
categories = ['A', 'B', 'C', 'D', 'E']
values = [5, 7, 3, 8, 6]
plt.bar(categories, values, color=['blue', 'green', 'red', 'purple', 'orange'])
plt.title("Bar Plot")
plt.show()
```

```python
data = np.random.randn(1000)
plt.hist(data, bins=30, color='green')
plt.title("Histogram")
plt.show()
```

```python
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [50, 25, 15, 10]
explode = (0.1, 0, 0, 0)
plt.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=140)
plt.title("Pie Chart")
plt.show()
```
